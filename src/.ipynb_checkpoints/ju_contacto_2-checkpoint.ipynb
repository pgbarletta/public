```julia
using CSV, Plots, Tables, DelimitedFiles, DataFrames, TableView, BrowseTables
using Statistics
data_dir = "/home/pbarletta/labo/21/p68/bis/data"
rst_dir = "/home/pbarletta/labo/21/p68/bis/rtdos_log"
rank_dir = "/home/pbarletta/labo/21/p68/bis/data/C8_ranking/C8_global_ranking"
```

```julia
steps_100 = convert(Array{Int64}, readdlm("steps_100")[:, 1])
func1_mut_100 = readdlm("func1_mut_100")[:, 1]
func2_mut_100 = readdlm("func2_mut_100")[:, 1]
s_100 = readdlm("srank_100")[:, 1]

steps_200 = convert(Array{Int64}, readdlm("steps_200")[:, 1])
func1_mut_200 = readdlm("func1_mut_200")[:, 1]
func2_mut_200 = readdlm("func2_mut_200")[:, 1]
s_200 = readdlm("srank_200")[:, 1]

steps_300 = convert(Array{Int64}, readdlm("steps_300")[:, 1])
func1_mut_300 = readdlm("func1_mut_300")[:, 1]
func2_mut_300 = readdlm("func2_mut_300")[:, 1]
s_300 = readdlm("srank_300")[:, 1];
```

```julia
avg_func1 = mean([ func1_mut_100 ; func1_mut_200 ; func1_mut_300 ])
avg_func1 *= .4
avg_func2 = mean([ func2_mut_100 ; func2_mut_200 ; func2_mut_300 ])
avg_srank = mean([ s_100 ; s_200 ; s_300 ])
```

```julia
top_mut_100 = (func1_mut_100 .> avg_func1) .&
    (func2_mut_100 .> avg_func2) .&  (s_100 .< avg_srank)

top_mut_200 = (func1_mut_200 .> avg_func1) .&
    (func2_mut_200 .> avg_func2) .& (s_200 .< avg_srank)

top_mut_300 = (func1_mut_300 .> avg_func1) .&
    (func2_mut_300 .> avg_func2) .& (s_300 .< avg_srank);
```

```julia
# .step is 0-indexed
df_top_mut_100 = DataFrame(step = steps_100[top_mut_100] .- 1,
    s = s_100[top_mut_100],
    func1 = func1_mut_100[top_mut_100],
    func2 = func2_mut_100[top_mut_100])

# .step is 0-indexed
df_top_mut_200 = DataFrame(step = steps_200[top_mut_200] .- 1,
    s = s_200[top_mut_200],
    func1 = func1_mut_200[top_mut_200],
    func2 = func2_mut_200[top_mut_200])

# .step is 0-indexed
df_top_mut_300 = DataFrame(step = steps_300[top_mut_300] .- 1,
    s = s_300[top_mut_300],
    func1 = func1_mut_300[top_mut_300],
    func2 = func2_mut_300[top_mut_300]);
```

```julia
scatter(func2_mut_100, s_100,
    lab = "", size=(800, 600),
    xlabel = "func2", ylabel = "S(rank)", colorbar_title = "func1",
    title = "mutants",
    color = :lightgrey, markerstrokewidth = 0, markersize = 5,
    xlims = (-10, 15), clims = (0, 10), ylims = (0, 1))

scatter!(func2_mut_200, s_200, lab = "",
    color = :lightgrey, markerstrokewidth = 0, markersize = 5)
scatter!(func2_mut_300, s_300, lab = "",
    color = :lightgrey, markerstrokewidth = 0, markersize = 5)

scatter!(df_top_mut_100.func2, df_top_mut_100.s, marker_z = df_top_mut_100.func1,
    lab = "", seriescolor = cgrad(:roma, rev = true),
    markerstrokewidth = 0, markersize = 8)
scatter!(df_top_mut_200.func2, df_top_mut_200.s, marker_z = df_top_mut_200.func1,
    lab = "", seriescolor = cgrad(:roma, rev = true),
    markerstrokewidth = 0, markersize = 8)
scatter!(df_top_mut_300.func2, df_top_mut_300.s, marker_z = df_top_mut_300.func1,
    lab = "", seriescolor = cgrad(:roma, rev = true),
    markerstrokewidth = 0, markersize = 8)

hline!([avg_srank], color = :deepskyblue, label = "avg_Srank")
vline!([avg_func2], color = :FireBrick, label = "avg_func2")
annotate!(16., .14, ("___", :black))
annotate!(14, .1, ("avg\n   func1", :black))
```

```julia
# df_top_mut_100.step is 0-indexed
count_top_mut_100 = length(df_top_mut_100.step)

plt_100ns = scatter(df_top_mut_100.func2, df_top_mut_100.s, marker_z = df_top_mut_100.func1,
    lab = "", size=(800, 600),
    xlabel = "func2", ylabel = "S(rank)", colorbar_title = "func1",
    title = "top 100ns mutants",
    seriescolor = cgrad(:roma, rev = true),
    markeralpha = 0.9, markerstrokewidth = 0, markersize = 8, 
    xlims = (3.5, 15), clims = (1.6, 10), ylims = (.0, avg_srank))

a = fill(+0.5, count_top_mut_100)
b = fill(-.01, count_top_mut_100)

scatter!(df_top_mut_100.func2 .+ a, df_top_mut_100.s .+ b,
    lab = :none,
    markersize = 0,
    series_annotations = string.(df_top_mut_100.step))
```
